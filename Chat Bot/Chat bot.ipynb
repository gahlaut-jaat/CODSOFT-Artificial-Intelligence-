```python
'''Import Libraries
   random - allow us to pick random choices/number
   re - its about pattern and text. Find match, pattern, replica, sequence'''

import random
import re

class ChatBot :
     negative_responsive = ("no","nope","nay","not a chance","sorry")
     exit_commands = ("quit","pause","exit","goodbye","bye")

     def __init__(self):
         self.bot_response = {
            'bot_details': r'.*\s*name.*',
            'state_of_bot': r'how\are\you.*',
            'user_residence': r'.*\s*residing.*',
            'about_INDIA': r'india.*',
            'about_population': r'.*\s*population.*',
            'about_education': r'.*literacy\education'
         }

     def greet(self):
         self.name = input("Hey! I am Buddy.\nWhat's your name?\n")
         will_help = input(f"Hello {self.name}, May, I assist you today?\n")
         if will_help in self.negative_responsive:
             print("Alright have a great day!")
             return
         self.chat()

     def make_exit(self, reply):
         for command in self.exit_commands:
             if command in reply:
                 print("Thanks for reaching out.\n Have a great day!\n")
                 return True
             return False

     def chat(self):
         reply = input("Please tell me your Query:").lower()
         while not self.make_exit(reply):
             reply = input(self.match_reply(reply))

     def match_reply(self, reply):
         for key, value in self.bot_response.items():
             intent = key
             regex_pattern = value
             found_match = re.match(regex_pattern, reply)
             if found_match and intent == 'bot_details':
                 return self.bot_details()
             elif found_match and intent == 'state_of_bot':
                 return self.state_of_bot()
             elif found_match and intent == 'user_residence':
                return self.user_residence()
             elif found_match and intent == 'about_INDIA':
                 return self.about_INDIA()
             elif found_match and intent == 'about_population':
                 return self.about_populatiion()
             elif found_match and intent == 'about_education':
                 return self.about_education()
         return  self.no_match_intent()

     def bot_details(self):
        responses = ("Hello, I am Buddy. I am a rule-based chatbot. I am here to help u out in any way.\n Thank you!")

     def state_of_bot(self):
        responses = ("I am fine. What's about you?")

     def user_residence(self):
        responses = ("Currently you are residing in INDIA")

     def about_INDIA(self):
        responses = ("India has a unique culture and is one of the oldest and greatest civilizations of the world.\nIndia covers an area of 32,87,263 sq. km, extending from the snow-covered Himalayan heights to the tropical rain forests of the south.\n As the 7th largest country in the world, India stands apart from the rest of Asia, marked off as it is by mountains and the sea, which give the country a distinct geographical entity.\nBounded by the Great Himalayas in the north, it stretches southwards and at the Tropic of Cancer, tapers off into the Indian Ocean between the Bay of Bengal on the east and the Arabian Sea on the west.\nIt has a land frontier of about 15,200 km. The total length of the coastline of the mainland, Lakshadweep Islands and Andaman & Nicobar Islands is 7,516.6 km.")

     def about_population(self):
        responses = ("Total Populatioin - 1,210,854,977\nMales Population - 623,724,568\nFemales Population - 586,469,294\n NOTE - This population record is as per 2011 population census")

     def about_education(self):
        responses = ("Total Literate - 74%\nMales Literate - 82.10%\nFemales Literate - 65.46%\n Kerla has the Highest Literacy Rate")

     def no_match_intent(self):
        responses = ("I'm sorry. I didn't quite understand that. Can you please rephrase?\n"
                     "My apologies, can you provide more details?\n")
        return random.choice(responses)

bot = ChatBot()
bot.greet()
            



    

```
