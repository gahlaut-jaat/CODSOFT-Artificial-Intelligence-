```python
import re

class IndiaChatBot:
    negative_responsive = ("no","nope","nay","not a chance","sorry")
    exit_commands = ("quit","pause","exit","goodbye","bye")
    def __init__(self):
        self.bot_response = {
            'bot_details': r'.*\s*who are you|who|name.*',
            'state_of_bot': r'.*\s*how are you.*',
            'country_details': r'.*\s*tell me about india|india.*',
            'population': r'.*\s*population.*',
            'culture': r'.*\s*culture|tradition.*',
            'geography': r'.*\s*geography|location.*',
            'history': r'.*\s*history.*'
        }

    def greet(self):
        self.name = input("Namaste! I am Buddy. Your guide to India.\n What's your name?\n")
        will_help = input(f"Hello {self.name}, May, I assist you today?\n")
        if will_help.lower() in self.negative_responsive:
             print("Alright have a great day!")
             return
        self.chat()

    def make_exit(self, reply):
         for command in self.exit_commands:
             if command in reply:
                 print("Thanks for reaching out.\n Have a great day!\n")
                 return True
         return False

    def chat(self):
        reply = input("Your question: ")
        while reply and not self.make_exit(reply):
            reply = input(self.match_reply(reply.lower()))

    def match_reply(self, reply):
        for key, value in self.bot_response.items():
            intent = key
            regex_pattern = value
            found_match = re.search(regex_pattern, reply)

            if found_match:
                if intent == 'bot_details':
                    return self.bot_details()
                elif intent == 'state_of_bot':
                    return self.state_of_bot()
                elif intent == 'country_details':
                    return self.country_details()
                elif intent == 'population':
                    return self.population()
                elif intent == 'culture':
                    return self.culture()
                elif intent == 'geography':
                    return self.geography()
                elif intent == 'history':
                    return self.history()

        return "Sorry, I don't have information on that. Ask me something else!\n"

    def bot_details(self):
        return "Hello, I am Buddy. I am a rule-based chatbot. I am here to help u out in any way.\n Thank you!\n"

    def state_of_bot(self):
        return "I am fine. What's about you?\n"

    def country_details(self):
        return "India, officially the Republic of India, is a country in South Asia.\nIt is the seventh-largest country by area and the most populous democracy in the world.\nIndia has a unique culture and is one of the oldest and greatest civilizations of the world.\nIndia covers an area of 32,87,263 sq. km, extending from the snow-covered Himalayan heights to the tropical rain forests of the south.\n As the 7th largest country in the world, India stands apart from the rest of Asia, marked off as it is by mountains and the sea, which give the country a distinct geographical entity.\nBounded by the Great Himalayas in the north, it stretches southwards and at the Tropic of Cancer, tapers off into the Indian Ocean between the Bay of Bengal on the east and the Arabian Sea on the west.\nIt has a land frontier of about 15,200 km. The total length of the coastline of the mainland, Lakshadweep Islands and Andaman & Nicobar Islands is 7,516.6 km.\nWant to know more?\n"

    def population(self):
        return "India has a population of over 1.21 billion people, making it the second-most populous country in the world.\nTotal Populatioin - 1,210,854,977\nMales Population - 623,724,568\nFemales Population - 586,469,294\n NOTE - This population record is as per 2011 population census.\n Anything else?\n"

    def culture(self):
        return "India is known for its rich and diverse culture, with a long history of art, music, dance, and literature.\n Interested in something specific?\n"

    def geography(self):
        return "India is located in South Asia, bordered by Pakistan, China, Nepal, Bhutan, Bangladesh, and Myanmar. It has a diverse landscape, including the Himalayas, the Ganges plain, and the Thar Desert.\n What else would you like to know?\n"

    def history(self):
        return "India has a long and fascinating history, dating back to the Indus Valley Civilization. It has been ruled by various empires and dynasties, including the Mughals and the British.\n Want to learn more about a specific period?\n"

bot = IndiaChatBot()
bot.greet()
```
